**The data preparation process will include the following:**<br>
1. Merged the external oil exporting countries name dataset to world_gdp dataset for unification <br>
2. Drop countries that are not oil exporting country from the data <br>
3. Maintain Data Integrity and ensure data quality assurance by<br>
    fixing data inaccuracy, <br>
    drop irrelevant column,<br>
    fix data incosistency

```python
# Import Necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import csv
```

```python
# Load the world gdp dataset as main_dt
gdp_dt = pd.read_csv("../data/processed/gdp_world_cleaned_v1.csv")
# load the oil exporting countries dataset as
oil_state_dt = pd.read_csv("../data/processed/oil_states_v1.csv")
```

```python
oil_state_gdp = pd.merge(oil_state_dt, gdp_dt, how="left", on="country_code", suffixes=(None, "_y"))
display(oil_state_gdp)
```

```python
oil_state_gdp_cleaned = oil_state_gdp.drop(columns=["country_name_y"])
display(oil_state_gdp_cleaned)
```

```python
# List of African oil-producing countries
african_countries = [
    "NGA",
    "LBY",
    "AGO",
    "DZA",
    "EGY",
    "COG",
    "GAB",
    "SDN",
    "TCD",
    "CMR",
    "GNQ",
    "TUN",
    "CIV",
    "COD",
]  # Add more African countries as needed

# Filter the DataFrame for African oil-producing countries
african_oil_producers = oil_state_gdp_cleaned[
    oil_state_gdp_cleaned["country_code"].isin(african_countries)
]

# Display African oil-producing countries
print("African Oil Producing Countries:")
print(african_oil_producers)

```

```python
# Export oil producing countries gdp dataset to csv file
oil_state_gdp_cleaned.to_csv(
    "../data/processed/oil_producer_gdp.csv",
    sep=",",
    quotechar='"',
    index=None, quoting=csv.QUOTE_ALL
) 
```

```python
import pandas as pd # for data manipulation
import csv # for csv file reading and exporting 
```

```python
countries_name_list = pd.read_csv("/home/ridwanullahi/Desktop/Work_Space/Play/oil-impact/data/processed/country/countries_list/world_countries_list.csv") # read country name list 
countries_name_list.info() # check the info of the data frame
countries_name_list.head(10) # check the first 10 rows of the data frame
# select row where country name is aruba from the country name list
#countries_name_list[countries_name_list['country_name'] == 'Aruba']
# sort the country name list by country code
#countries_name_list_v1 = countries_name_list.sort_values(by=['country_code'])
#countries_name_list_v1.head(10) # check the first 10 rows of the data frame
```

```python
gdp_world = pd.read_csv("/home/ridwanullahi/Desktop/Work_Space/Play/oil-impact/data/processed/gdp/world/gdp_world_cleaned_v1.csv")
gdp_world.head(10) # check the head of the data frame
```

```python
# merge the country name list with the gdp world data frame together on the country code column to validate and ensure correct countries name in the column column from the gdp countries list
gdp_world_v1 = pd.merge(gdp_world, countries_name_list, on='country_code', how='inner', suffixes= (None, "_y"))
gdp_world_v1.head(10) # check the head of the data frame
```

```python
gdp_world_v2 = gdp_world_v1.drop(['country_name_y'], axis=1) # drop the column named country_name_y
gdp_world_v2.head(10) # check the head of the data frame
```

```python
# create a column to indicate the oil exporting countries and non oil exporting countries
# first create a list of oil exporting countries
oil_countries_list = pd.read_csv("/home/ridwanullahi/Desktop/Work_Space/Play/oil-impact/data/processed/country/countries_list/oil_states.csv")

#convert the oil countries list to a list
oil_countries_list_v1 = oil_countries_list['country_name'].tolist()
oil_countries_list_v1
```

```python
# create a column named state_indicator to indicate the oil exporting countries and non oil exporting countries 
gdp_world_v2['state_indicator'] = gdp_world_v2['country_name'].apply(lambda x: 'Oil Producer' if x in oil_countries_list_v1 else 'Non oil Producer')
gdp_world_v2.describe(include="all") # check the head of the data frame
```

```python
# save the data frame to a csv file
gdp_world_v2.to_csv("/home/ridwanullahi/Desktop/Work_Space/Play/oil-impact/data/processed/gdp/world/gdp_world_final_insight_03.csv",quotechar='"', quoting=csv.QUOTE_ALL, sep=",", index=False)
```

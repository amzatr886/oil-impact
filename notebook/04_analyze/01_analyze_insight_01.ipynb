```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# Load oil producing countries dataset 
oil_exporters  = pd.read_csv("../data/processed/oil_producer_gdp.csv")
oil_exporters.describe(include="all")
oil_exporters.head()
```

```python
# Create a column named continent to identify africa countries from others 

african_countries = ["Nigeria",
"Libya",
"Angola",
"Algeria",
"Egpyt",
"Republic of congo",
"Gabon",
"Sudan",
"Chad","Cameroon",
"Equatorial guinea"
,"Tunisia",
"Ivory cost",
"Democratic republic of Congo"]

# Create the 'Continent' column
oil_exporters['Continent'] = oil_exporters['country_name'].apply(lambda x: 'Africa' if x in african_countries else 'Other')
oil_exporters.describe(include="all")
```

```python
oil_exporters.head()
```

```python
# Convert numeric columns to appropriate data type
numeric_columns = [str(x) for x in range(2000, 2023)]
# oil_exporters[numeric_columns] = oil_exporters[numeric_columns].astype(float)

# Perform mean aggregation
african_vs_others = oil_exporters.groupby('Continent')[numeric_columns].mean().reset_index()
african_vs_others.head()
```

```python
# Adjust the african_vs_others dataframe to be in the right format for plotting
# Ploting GDP against year for African countries and other countries
# Convert the columns to rows.
african_vs_others_transposed = african_vs_others.transpose().reset_index()

# Rename the columns
african_vs_others_transposed.columns = african_vs_others_transposed.iloc[0]
african_vs_others_transposed = african_vs_others_transposed.drop(0)

# Rename the index column
african_vs_others_transposed = african_vs_others_transposed.rename(columns={'Continent': 'Year'})

# Convert the year column to numeric
african_vs_others_transposed['Year'] = african_vs_others_transposed['Year'].astype(int)

# Reseting the index
african_vs_others_transposed = african_vs_others_transposed.reset_index(drop=True)

# Transform data to billions
african_vs_others_transposed['Africa'] = african_vs_others_transposed['Africa'].apply(lambda x: round(x/1e9, 2))
african_vs_others_transposed['Other'] = african_vs_others_transposed['Other'].apply(lambda x: round(x/1e9, 2))

african_vs_others_transposed.head()
```

```python
sns.lineplot(data=african_vs_others_transposed, x="Year", y="Africa", label="Africa", color="green", legend=True)
sns.lineplot(data=african_vs_others_transposed, x="Year", y="Other", label="Other", color="blue", legend=True)

plt.title("GDP of African countries vs other countries (in billions)")
plt.ylabel("GDP (in billions)")
plt.xlabel("Year")
plt.xticks(african_vs_others_transposed['Year'].unique().tolist(), rotation=90)
plt.grid(True)  # Add grid
plt.show()
```

```python
# Ploy year vs GDP for African countries and other countries
sns.lineplot(data=african_vs_others_transposed)

# Add legend, title, x and y labels
plt.title("GDP of African countries vs other countries (in billions)")
plt.ylabel("GDP (in billions)")
plt.xlabel("Year")
plt.show()
```

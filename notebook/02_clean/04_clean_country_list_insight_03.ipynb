```python
import pandas as pd #for data cleaning
import csv # for exporting data to csv 
```

```python
# Read in the data
gdp = pd.read_csv("/home/ridwanullahi/Desktop/Work_Space/Play/oil-impact/data/raw/world-gdp/world_gdp.csv")
gdp.info()
gdp.head()
```

```python
# rename (country name to country_name, country code to country_code, indicator name to indicator_name, indicator code to indicator_code)
gdp_v1 = gdp.rename(columns={"Country Name": "country_name", "Country Code": "country_code", "Indicator Name": "indicator_name", "Indicator Code": "indicator_code"})
gdp_v1.head()
```

```python
# drop  columns (Unnamed:67, indicator_code)
gdp_v2 = gdp_v1.drop(columns=['Unnamed: 67', 'indicator_code'])
gdp_v2.head()
```

```python
# drop rows that are less than 2000
gdp_v3 = gdp_v2.drop(gdp_v2.columns[3:43], axis=1)
gdp_v3.head()
gdp_v3.info()
```

```python
# import country list dataset
world_countries_list = pd.read_json("/home/ridwanullahi/Desktop/Work_Space/Play/oil-impact/data/raw/countries/country.json")
headers = world_countries_list.iloc[0].values # select the first row of the dataframe and extract the values as numpy array
world_countries_list.columns = headers # set the header of dataframe rows as the extracted numpy array
world_countries_list.drop(index=0, axis=0, inplace=True) # drop the first row of the dataframe
```

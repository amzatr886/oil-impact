**The Cleaning Process include The following:**<br>
Remove duplicated  value in the reqiured cell <br>
Rename some column to short and descriptive name <br>
Drop irrelevant column to our analysis 



```python
# Import necessary libraries
import pandas as pd
import csv
```

```python
world_gdp = pd.read_csv("/home/ridwanullahi/Desktop/Work_Space/Play/oil-impact/data/raw/world-gdp/world_gdp.csv")
```

```python
display(world_gdp.head(10))
```

```python
# Rename Column to short and descriptive name
world_gdp_renamed = world_gdp.rename(
    columns={
        "Country Name": "country_name",
        "Country Code": "country_code",
        "Indicator Name": "currency",
        "Indicator Code": "indicator_code",
    }
)
```

```python
display(world_gdp_renamed.head(10))
```

```python
# Drop irrelevant columns to our analysis
world_gdp_v1 = world_gdp_renamed.drop(
    columns=[
        "1960",
        "1961",
        "1962",
        "1963",
        "1964",
        "1965",
        "1966",
        "1967",
        "1967",
        "1968",
        "1969",
        "1970",
        "1971",
        "1972",
        "1973",
        "indicator_code",
        "Unnamed: 67",
    ]
)
display(world_gdp_v1.head(10))
```

```python
# drop rows before 2000
selected_columns = ["country_name", "country_code", "currency"] + [
    str(year) for year in range(2000, 2023)
]
gdp_world_2000_2022_v1 = world_gdp_renamed.loc[:, selected_columns]
```

```python
# Export the dataframe to csv file
gdp_world_2000_2022_v1.to_csv(
    "../data/processed/gdp_world_cleaned_v1.csv",
    sep=",",
    quotechar='"',quoting=csv.QUOTE_ALL,
    index=None,
)
```

**The Data Cleaning Process Include:** <br>
1. Drop Duplicate <br>
2. Remove White space <br>
3. Drop null cell <br>
4. Fix data inaccuracy <br>

```python
# Import necessary library
import pandas as pd
```

```python
oil_states = pd.read_csv("../data/raw/world-gdp/oil_producing_state.csv")
display(oil_states)
oil_states.describe(include="all")
```

```python
# Remove duplicated Value
oil_states_unique = oil_states.drop_duplicates().reset_index(drop=True)
display(oil_states_unique)
oil_states_unique.describe(include="all")
```

```python
# Drop irrelevant column to our analysis and fix data inaccuarcy
oil_states_unique_clean = oil_states_unique.drop(columns=["Unnamed: 1"])
display(oil_states_unique_clean)
oil_states_unique_clean.describe(include="all")
```

```python
oil_states_final = (
    oil_states_unique_clean.rename(columns={"States": "country_name"})
    .sort_values("country_name")
    .reset_index(drop=True)
)
display(oil_states_final)
```

```python
# Export dataframe as csv file
oil_states_final.to_csv("../data/processed/oil_states.csv", index=False)
```

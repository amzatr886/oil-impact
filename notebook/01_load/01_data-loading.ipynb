```python
oil_states_no_dup = oil_states.drop_duplicates().reset_index(drop=True)
```

```python
display(oil_states_no_dup)
```

```python
oil_states_clean = oil_states_no_dup.drop(columns=["Unnamed: 1"])
```

```python
oil_states_clean.describe(include="all")
```

```python
display(oil_states_clean)
```

```python
oil_states_clean = (
    oil_states_clean.rename(columns={"States": "Country"})
    .sort_values(by="Country")
    .reset_index(drop=True)
)
display(oil_states_clean)
```

```python
import pandas as pd

# Load the dataset
gdp_df = pd.read_csv('../data/processed/gdp_world_cleaned_v1.csv')

# Define a list of African countries
african_countries = ["Nigeria",
"Libya",
"Angola",
"Algeria",
"Egpyt",
"Republic of congo",
"Gabon",
"Sudan",
"Chad","Cameroon",
"Equatorial guinea"
,"Tunisia",
"Ivory cost",
"Democratic republic of Congo"]

# Create the 'Continent' column
gdp_df['Continent'] = gdp_df['country_name'].apply(lambda x: 'Africa' if x in african_countries else 'Other')
print(gdp_df)
```
